### **Commits**

SHA: Es el identificador unico del commit.
 Author: 
 El autor del commit.
 Commit message(Fix):
 El mensaje del commit.

Cambiar mensaje del commit:
 
 git commit --amend

 

 

Como deshacer el ultimo commit

git reset --soft HEAD~1 

Solo se quitara el commit mas nos los cambios del archivo.



Podemos usar **hard** para que no se guarden los cambios del archivo.

### **RAMAS**

- Para crear ramas: git branch nombre_rama
- Para ver todas las ramas: git branch





- Para cambiar el nombre_rama: git branch -m





- Para cambiar de rama: git checkout nombre_rama



- Para fusionar ramas: git merge nombre_rama


- Para eliminar una rama: git branch -d nombre_rama
